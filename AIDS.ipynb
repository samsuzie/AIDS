```python
import scipy.stats as stats

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.distplot(df1['time'])
plt.title('time distribution')

plt.subplot(1,2,2)
stats.probplot(df1['time'],dist='norm',plot = plt)
plt.title("time plot")

plt.show()
```

```python
# now will calculate outlier
percentile_25 = df1['time'].quantile(0.25)
percentile_75 = df1['time'].quantile(0.75)

# inter-quartile range
iqr = percentile_75-percentile_25
iqr
```

```python
# finding the outlier
percentile25 = df1["time"].quantile(0.25)
percentile75 = df1["time"].quantile(0.75)

iqr = percentile75-percentile25
iqr
# checking upper and lower limit in distance variable

upper_limit = percentile75 + 1.5*iqr
lower_limit = percentile25 - 1.5*iqr

print("upper_limit",upper_limit)
print("lower_limit",lower_limit)


# capping
new_df_cap = df1.copy()

new_df_cap["time"] = np.where(
    new_df_cap["time"]>upper_limit,
    upper_limit,
    np.where(
       new_df_cap["time"] < lower_limit,
       lower_limit,
       new_df_cap["time"]
    )
)

df1["time"] = new_df_cap["time"]
```

```python
X_new = df1.drop('infected',axis=1)
```

```python
y_new = df1["infected"]
```

```python
# now scale the dataframe
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
pre_process = MinMaxScaler().fit(X_new)
x_transform = pre_process.transform(X_new)
```

```python
# now split the data in training and gtesting to perform train test split
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_transform, y_new, test_size=0.2, random_state=101)
```

```python
# import the model 
from sklearn.linear_model import LogisticRegression
```

```python
lgr = LogisticRegression()
```

```python
lgr.fit(X_train,y_train)
```

```python
pred=lgr.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,pred))
print(confusion_matrix(y_test,pred))
```

```python
from sklearn.metrics import accuracy_score
pred = lgr.predict(X_test)

# score it
lgr_accuracy = round(accuracy_score(y_test,pred)*100,2)
print("accuracy",lgr_accuracy,'%')
```

```python
# accuracy is 86.92 will try to improve accuracy by using desicion tree 
```

```python
from sklearn.metrics import accuracy_score
pred_tree = dtc.predict(X_test)

# score it

dtc_accuracy = round(accuracy_score(y_test,pred_tree)*100,2)
print("accuracy",dtc_accuracy,'%')
```
